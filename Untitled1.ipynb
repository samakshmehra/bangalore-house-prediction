```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Bengaluru_House_Data.csv")
```

```python
data["location"] =data["location"].fillna('Sarjapur  Road ')
```

```python
data["size"] =data["size"].fillna('2 BHK')
```

```python
data["bath"] =data["bath"].fillna(data["bath"].median())
```

```python
data["bhk"] = data['size'].str.split().str.get(0).astype(int)
```

```python
def  convertrang(x) :
    temp =x.split('-')
    if len(temp) ==2 :
        return(float(temp[0]) + float(temp[1]))/2

    try :
        return float(x)
    except:
        return None
```

```python
data.info()
```

```python
data["total_sqft"] =data["total_sqft"].apply(convertrang)
```

```python
data.drop(columns=['area_type','availability','balcony','society'],inplace=True)
```

```python
data.head()
```

```python
data['price_per_sqft'] =data ["price"] *100000 /data["total_sqft"] 
```

```python
data.describe()
```

```python
data["location"] = data["location"].apply(lambda x: x.strip())
```

```python
location_count = data["location"].value_counts()
```

```python
location_count_less_10 = location_count[location_count<=10]
data["location"]=data["location"].apply( lambda x:'others' if x in location_count_less_10 else x)
```

```python
data = data[((data['total_sqft']/data["bhk"])>=300)]
data.describe()
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()  # Corrected typo: DataFrame instead of Dataframe
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        gen_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_output = pd.concat([df_output, gen_df], ignore_index=True)
    return df_output

data = remove_outliers_sqft(data)
data.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'): 
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'): 
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats["count"] > 5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')

# Apply the function to your DataFrame
data = bhk_outlier_remover(data)
```

```python
data = bhk_outlier_remover(data)
```

```python
x=data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python

column_trans = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), ['location'])],
    remainder='passthrough'
)
```

```python
# Create the pipeline
scaler = StandardScaler(with_mean=False)  # Set with_mean=False for sparse matrices
lr = LinearRegression()
pipe = make_pipeline(column_trans, scaler, lr)

# Fit the pipeline
pipe.fit(x_train, y_train)
```

```python
y_predict_linear = pipe.predict(x_test)
r2_score(y_test,y_predict_linear)
```
